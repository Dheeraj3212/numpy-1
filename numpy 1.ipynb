```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def get_intersection_node(headA, headB):
    if not headA or not headB:
        return None
    
    # Pointers for both lists
    pointerA = headA
    pointerB = headB
    
    # Traverse both lists until they meet or reach the end
    while pointerA != pointerB:
        # Move pointers to the next node
        pointerA = pointerA.next if pointerA else headB
        pointerB = pointerB.next if pointerB else headA
    
    # Return the intersection node
    return pointerA

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Create the linked lists
list1 = ListNode(1)
list1.next = ListNode(2)
list1.next.next = ListNode(3)
list1.next.next.next = ListNode(4)

list2 = ListNode(9)
list2.next = ListNode(8)
list2.next.next = list1.next.next

print("Input:")
print("List 1: ", end="")
print_linked_list(list1)
print("List 2: ", end="")
print_linked_list(list2)

# Find the intersection node
intersection_node = get_intersection_node(list1, list2)

if intersection_node:
    print("Output: Node with value", intersection_node.val)
else:
    print("Output: No intersection found")
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def add_two_numbers(l1, l2):
    dummy = ListNode()
    current = dummy
    carry = 0
    
    while l1 or l2 or carry:
        # Calculate sum and carry
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0
        total = val1 + val2 + carry
        carry = total // 10
        digit = total % 10
        
        # Create new node with the sum digit
        current.next = ListNode(digit)
        current = current.next
        
        # Move to the next nodes
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None
    
    return dummy.next

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Create the linked lists
list1 = ListNode(2)
list1.next = ListNode(4)
list1.next.next = ListNode(3)

list2 = ListNode(5)
list2.next = ListNode(6)
list2.next.next = ListNode(4)

print("Input:")
print("List 1: ", end="")
print_linked_list(list1)
print("List 2: ", end="")
print_linked_list(list2)

# Add two numbers
sum_list = add_two_numbers(list1, list2)

print("Output: ", end="")
print_linked_list(sum_list)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def swap_pairs(head):
    dummy = ListNode(0)
    dummy.next = head
    prev = dummy
    
    while prev.next and prev.next.next:
        first = prev.next
        second = prev.next.next
        
        # Swap nodes
        prev.next = second
        first.next = second.next
        second.next = first
        
        # Move to the next pair
        prev = first
    
    return dummy.next

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Create the linked list
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)

print("Input: ", end="")
print_linked_list(head)

# Swap nodes in pairs
new_head = swap_pairs(head)

print("Output: ", end="")
print_linked_list(new_head)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverse_k_group(head, k):
    if not head or k == 1:
        return head
    
    dummy = ListNode(0)
    dummy.next = head
    prev = dummy
    
    while True:
        # Check if there are at least k nodes remaining
        tail = prev
        for _ in range(k):
            tail = tail.next
            if not tail:
                return dummy.next
        
        # Reverse the k nodes
        next_group = tail.next
        prev.next = reverse_sublist(prev.next, tail)
        
        # Stitch the reversed sublist back to the original list
        prev = tail
        prev.next = next_group
    
    return dummy.next

def reverse_sublist(head, tail):
    prev = None
    current = head
    
    while prev != tail:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    
    return prev

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Create the linked list
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Input: ", end="")
print_linked_list(head)

# Reverse nodes in groups of k
k = 3
new_head = reverse_k_group(head, k)

print("Output: ", end="")
print_linked_list(new_head)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def is_palindrome(head):
    if not head or not head.next:
        return True
    
    # Find the middle of the linked list
    slow = head
    fast = head
    
    while fast.next and fast.next.next:
        slow = slow.next
        fast = fast.next.next
    
    # Reverse the second half of the linked list
    second_half_head = reverse_linked_list(slow.next)
    
    # Compare the first half with the reversed second half
    while second_half_head:
        if head.val != second_half_head.val:
            return False
        head = head.next
        second_half_head = second_half_head.next
    
    return True

def reverse_linked_list(head):
    prev = None
    current = head
    
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    
    return prev

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Create the linked list
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(2)
head.next.next.next = ListNode(1)

print("Input: ", end="")
print_linked_list(head)

# Determine if the linked list is a palindrome
result = is_palindrome(head)

print("Output:", result)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotate_right(head, k):
    if not head or k == 0:
        return head
    
    # Calculate the length of the linked list
    length = 1
    tail = head
    while tail.next:
        tail = tail.next
        length += 1
    
    # Calculate the actual rotation value
    k = k % length
    if k == 0:
        return head
    
    # Move to the (length - k)th node from the beginning
    count = length - k
    current = head
    while count > 1:
        current = current.next
        count -= 1
    
    # Adjust the pointers to rotate the list
    new_head = current.next
    current.next = None
    tail.next = head
    
    return new_head

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Create the linked list
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Input: ", end="")
print_linked_list(head)

# Rotate the linked list to the right by k places
k = 2
new_head = rotate_right(head, k)

print("Output: ", end="")
print_linked_list(new_head)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rearrange_linked_list(head):
    if not head or not head.next:
        return head
    
    odd_head = ListNode(0)
    even_head = ListNode(0)
    odd_current = odd_head
    even_current = even_head
    current = head
    is_odd = True
    
    while current:
        if is_odd:
            odd_current.next = current
            odd_current = odd_current.next
        else:
            even_current.next = current
            even_current = even_current.next
        
        current = current.next
        is_odd = not is_odd
    
    odd_current.next = None
    even_current.next = None
    
    # Merge the two lists
    odd_current.next = even_head.next
    
    return odd_head.next

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Create the linked list
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Input: ", end="")
print_linked_list(head)

# Rearrange the linked list
new_head = rearrange_linked_list(head)

print("Output: ", end="")
print_linked_list(new_head)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def add_one(head):
    dummy = ListNode(0)
    dummy.next = head
    last_non_nine = dummy
    current = head
    
    # Find the last non-nine node
    while current:
        if current.val != 9:
            last_non_nine = current
        current = current.next
    
    # Increment the last non-nine node by one
    last_non_nine.val += 1
    
    # Set all nodes after the last non-nine node to zero
    current = last_non_nine.next
    while current:
        current.val = 0
        current = current.next
    
    # If the first node is now 10, update the head
    if dummy.val == 10:
        dummy.val = 1
        return dummy
    else:
        return dummy.next

# Function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Create the linked list
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)

print("Input: ", end="")
print_linked_list(head)

# Add one to the linked list
new_head = add_one(head)

print("Output: ", end="")
print_linked_list(new_head)
```

```python
def search_insert(nums, target):
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    # If target is not found, return the insertion position
    return left

# Test the function
nums = [1, 3, 5, 6]
target = 5
print("Input: nums =", nums, ", target =", target)
print("Output:", search_insert(nums, target))
```
